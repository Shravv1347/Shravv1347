Step 1- Data Cleaning

```python
import pandas as pd


df = pd.read_csv("/content/tinatic.csv")


print("Missing values:\n", df.isnull().sum())
```

```python
df.drop(columns=['Cabin'], inplace=True)


df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


print("Cleaned dataset info:")
print(df.info())
```

```python
print("Missing values:\n", df.isnull().sum())
```

Step 2- Exploratory Data Analysis (EDA)

1. Survival Rate by Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Count by Gender')
plt.show()
```

2. Survival Rate by Passenger Class


```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Count by Passenger Class')
plt.show()
```

3. Age Distribution

```python
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

4. Fare distribution by Survival

```python
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare vs Survival')
plt.show()
```

**Patterns & Trends to Highlight**

Women had a higher survival rate than men.

First-class passengers were more likely to survive than those in 2nd or 3rd class.

Younger passengers (children) had higher survival rates.

Survivors paid higher fares on average, indicating a correlation between wealth/class and survival.
