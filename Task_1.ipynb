Task 1 - Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "/content/population_by_age_group.csv"
df = pd.read_csv(file_path)


age_groups = ['0-4 years', '5-14 years', '15-24 years', '25-64 years', '65+']


age_distribution = df[age_groups].sum()


plt.figure(figsize=(10, 6))
plt.bar(age_distribution.index, age_distribution.values, color='red', edgecolor='black')
plt.title('Global Population Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Population in Billions')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
